```python
for i in range(1,5):
    fila= ""
    
    for j in range(1,i+1):
        fila=fila+str(i)
    print(fila)
    
```

```python
for i in range(1,5):
    fila= "*"
    
    for j in range(i):
        fila=fila+"*"
    print(fila)
```
