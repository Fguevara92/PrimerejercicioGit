```python
# --- DATOS DE ENTRADA ---
edad_txt = "16"
dinero_txt = "45.50"
precio_entrada_txt = "40.00"

viene_con_adulto = True
lista_negra = False

edad = int(edad_txt)
dinero = float(dinero_txt)
precio_entrada = float(precio_entrada_txt)

dinero_sobrante = dinero - precio_entrada

condicion_edad = ((edad >= 18) or viene_con_adulto)

condicion_dinero = (dinero_sobrante >= 0)

condicion_lista = not lista_negra

compra_aprobada = condicion_edad and condicion_dinero and condicion_lista


print("---RESULTADO DE LA COMPRA---")
print("Dinero Sobrante:",dinero_sobrante)
print("Compra Aprobada:", compra_aprobada)
```
