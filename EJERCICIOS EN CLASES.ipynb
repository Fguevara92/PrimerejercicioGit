```python
#Tipos de Datos

#String
nombre = str(input("Ingrese su nombre: "))
#INT - Entero
edad = int(input("Ingrese su edad: "))
#FLOAT - Decimal
altura = float(input("Ingrese su altura en centimetros: "))

#BOOLEAN - True o False

tiene_cedula = bool(int(input("Ingrese 1 (si) o 0 (no): ")))
 

print("\n--- Datos ---")
print("Nombre:", nombre, "\nEdad:", edad, "\nAltura:", altura, "\n¿Tiene cedula?:", tiene_cedula )
```

```python
#Clasificador de edad
edad = int(input("Ingrese la edad: "))

if edad < 12:
    print("Niño/Niña")
elif edad <= 18:
    print("Adolescente")
elif edad < 65:
    print("Adulto/Adulta")
else:
    print("Adulto mayor")
```
