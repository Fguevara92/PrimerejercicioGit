```python
#Programa Evaluador de Creditos
#Objetivo:Demostrar la estructura de un arbol de decision anidado
 
#Score Crediticio
for score_cliente in range(400,800,50):
    deuda_mensual = 400
    ingreso_mensual= 2000
    #Proporcion de ingresos comprometidos
    relacion_deuda=deuda_mensual / ingreso_mensual
    print("Procesando cliente - Score", score_cliente ,"Relacion deuda",relacion_deuda)
    #Nodo Raiz Clientes con historial excelente
    if score_cliente >= 700:
        #Rama A
        if relacion_deuda <= 0.35:
            print("Aprobado Automaticamente")
        else:
            print("Buen score sin embargo se encuentra muy endeudado")
    #Nodo Secundario Clientes con historial moderado
    elif score_cliente >=500:
        #Rama B
        if relacion_deuda <= 0.20:
            print("Revision manual por parte del personal","\n","Score medio, bajo riesgo de deuda")
        else:
            print("Score medio y alta deuda")
    else:
        print("Rechazado Automaticamente, Riesgo financiero critico")
```

```python
#EJER. PIZZA
usuario="0"
print("--- BIENVENIDO A PIZZABOT ---")
print("(Escribe 'salir' para terminar la conversación)")

while True:

    mensaje_cliente = input("\n", usuario,": Hola, soy PizzaBot. ¿En qué te puedo ayudar?: ")
    mensaje_cliente = mensaje_cliente.lower()

    
    if "pedido" in mensaje_cliente or "comprar" in mensaje_cliente or "menu" in mensaje_cliente or "menú" in mensaje_cliente:
        if "vegana" in mensaje_cliente:
            print("PizzaBot: Tenemos la Pizza Huerto por $12. ¿Deseas pedirla?")
        else:
            print("PizzaBot: Aquí tienes nuestro menú clásico: Pepperoni, Hawaiana y Queso. ¿Cuál prefieres?")

    
    elif "estado" in mensaje_cliente or "dónde" in mensaje_cliente or "donde" in mensaje_cliente or "demora" in mensaje_cliente:
        print("PizzaBot: Tu pedido ya salió del horno y llegará en aproximadamente 15 minutos.")
    
    
    elif "queja" in mensaje_cliente or "fría" in mensaje_cliente or "fria" in mensaje_cliente or "mal" in mensaje_cliente:
        print("PizzaBot: Lamentamos el inconveniente. Te transferimos con el administrador de turno.")

    elif "salir" in mensaje_cliente:
        print("PizzaBot: ¡Gracias por contactarnos! Apagando sistema...")
        break 
    
    
    else:
        print("PizzaBot: Aún estoy aprendiendo. Por favor, indica si deseas hacer un pedido, consultar un estado o presentar una queja.")
```
