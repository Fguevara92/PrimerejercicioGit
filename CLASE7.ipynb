```python
#Clase 7 - Bucles(For-While)
#While - Primero determina la condicion y si es verdadera la repite, caso contrario termina el bucle
opcion = 0

while (opcion != 4):
    print("----Bienvenidos al Menu----")
    print("1.Calculadora con dos numeros")
    print("2.Determinar numeros pares e impares")
    print("3.Sucesión de Fibonacci")
    print("4.Salir del programa")
    opcion = int(input("Ingrese la opcion deseada:"))
    if opcion == 1 :
        print("---Calculadora--")
        numero_uno = float(input("Ingrese el primer numero"))
        numero_dos = float(input("Ingrese el segundo numero"))
        suma = numero_uno + numero_dos
        resta = numero_uno - numero_dos
        mul = numero_uno * numero_dos
        div = numero_uno / numero_dos
        print("Suma es",suma,"\n",
              "Resta es:",resta,"\n",
              "Multiplicacion es:",mul,"\n",
              "Division:",div)
    elif opcion == 2:
        #Pedir al usuario que ingrese un numero, mediante ese numero ustedes deben determinar si es par o impar.
        print("---Números Pares e Impares---")
        numero = int(input("Ingrese un número entero: "))
        if numero % 2 == 0:
            print("El número", numero, "es PAR.")
        else:
            print("El número", numero, "es IMPAR.")  

    elif opcion == 3:
        #Sucesión de Fibonacci
        a=int(input("Ingrese numero uno: "))
        b=int(input("Ingrese numero dos: "))
        rango=int(input("ingrese numero rango: "))

        for i in range(rango):
            print(a)
            temporal = a + b
            a = b 
            b = temporal
            
    elif opcion == 4:
        print("Finalizar Programa")
    else:
        print("Opcion Invalida. Selecciona el numero 1, 2, 3, 4")
```
