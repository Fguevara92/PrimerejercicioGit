```python
##### Variables iniciales del sistema
opcion = 0
total_cuenta = 0.00
vuelto=0.00
descuento_aplicado = False

while opcion != 5:
    print("\n*** BIENVENIDO A LA CAJA REGISTRADORA ***")
    print("1. Agregar Encebollado ($3.00)")
    print("2. Agregar Empanada ($1.50)")
    print("3. Aplicar Descuento (10%)")
    print("4. Ver Cuenta Actual")
    print("5. Pagar y Salir")
    
    opcion = int(input("Elige una opción: "))
    
    if opcion == 1:
        #Encebollado
        cantidad = int(input("¿Cuántas unidades?: "))
        total_cuenta = total_cuenta + (cantidad * 3.00)
        print("-->", cantidad, "Encebollado(s) agregado(s).")
        
    elif opcion == 2:
        #Empanada
        cantidad = int(input("¿Cuántas unidades?: "))
        total_cuenta = total_cuenta + (cantidad * 1.50)
        print("-->", cantidad, "Empanada(s) agregada(s).")
        
    elif opcion == 3:
        #Descuento
        if descuento_aplicado == True:
            print("--> Error: El descuento ya fue aplicado en esta compra.")
        elif total_cuenta <= 10.00:
            print("--> Error: La cuenta debe ser mayor a $10.00 para aplicar el descuento.")
        else:
            #aplicacion del descuento
            total_cuenta = total_cuenta * 0.90
            descuento_aplicado = True 
            print("--> ¡Descuento del 10% aplicado con éxito!")
            
    elif opcion == 4:
        # cuenta
        print("Su cuenta actual es: $", total_cuenta)
        
    elif opcion == 5:
        #pagar y salir
        print("Total a cobrar: $", total_cuenta)
        efectivo = float(input("¿Efectivo recibido?: "))
        
        if efectivo >= total_cuenta:
            vuelto = efectivo - total_cuenta
            print("Pago exitoso. Su vuelto es: $", vuelto)
            print("¡Gracias por su compra, cerrando sistema!")
        else:
            print("--> Fondos insuficientes. Pago denegado.")
            
            opcion = 0 
            
    else:
        print("--> Opción inválida. Seleccione un número del 1 al 5.")
```
