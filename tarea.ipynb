```python
contrasena = input("Contraseña: ")

long_contra = len(contrasena) >= 8
mayuscula = False
numero = False

for caracter in contrasena:
    if caracter.isupper():
        mayuscula = True
    elif caracter.isdigit():
        numero = True

if long_contra and mayuscula and numero:
    print("Su contraseña es segura")
else:
    print("Su contraseña le falta seguridad:")

    if not long_contra:
        print("- No cumple los 8 caracteres")
    if not mayuscula:
        print("- No cumple al menos una mayuscula")
    if not numero:
        print("- No cumple al menos un numero")
```
